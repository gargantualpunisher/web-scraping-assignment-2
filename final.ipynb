```python
# Q1

driver = webdriver.Chrome(r"C:\Users\Dell\Videos\chromedriver_win32\chromedriver.exe")

time.sleep(10)

driver.get('https://www.naukri.com/')


# designation

designation = driver.find_element(By.CLASS_NAME,"suggestor-input")

designation.send_keys('Data Analyst')


#location 

designation = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")

designation.send_keys('Banglore')

time.sleep(5)
# search 

search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()

time.sleep(5)
# store

job_title =[]
job_location = []
company_name = []
experience_required = []

# job_title

title_tags = driver.find_elements(By.XPATH, '//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)

time.sleep(5)    

# location

location = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft locWdth"]')
for i in location[0:10]:
    loca =i.text
    job_location.append(loca)
    


# company_name

name = driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for i in name[0:10]:
    Name =i.text
    company_name.append(Name)




# Experience



experi = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft expwdth"]')
for i in experi[0:10]:
    Name_ex = i.text
    experience_required.append(Name_ex)
time.sleep(5)    
import pandas as pd

final = pd.DataFrame()

final['Job_Title']= job_title
final['Job_Title']= job_location
final['Company_Name']= company_name
final['Experience']= experience_required

final
```
