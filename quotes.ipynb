```python
from selenium import webdriver
from time import sleep
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.service import Service
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
```

```python
driver = webdriver.Chrome('D:\selenium\chromedriver.exe')
```

```python
driver.get('https://www.azquotes.com/')
```

```python
top = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top.click()
```

```python
Quote =[]
Author =[]
Type_of_Quote =[]
```

```python
# quote 

write = driver.find_elements(By.CLASS_NAME, "title")

for i in write:
    Quote.append(i.text)
    
len(Quote)
```

```python
# AUTHOR 

author = driver.find_elements(By.LINK_TEXT, "author")

for i in author:
    
    Author.append(i)

Author
```

```python
# type
type = driver.find_elements(By.CLASS_NAME, "tags")

for i in type:
    Type_of_Quote.append(i.text)
```
